## **Why RF-DETR? "NMS-free" (Non-Maximum Suppression) architecture.**
- It treats every object as unique, which is critical in satellite and aerial imagery where small, overlapping objects (like cars or storage tanks) often get "erased" by older CNN-based models.

## **Strawberry Ripeness Detection**

```python
from rfdetr import RFDETRMedium
import torch

# 1. Initialize the SOTA Medium model
model = RFDETRMedium(model_type='rfdetr-medium', 
                     resolution=576, 
                     accept_platform_model_license=True)

# 2. Start Fine-Tuning 
# Keep total_batch_size (batch * accum) = 16 for stability
model.train(
    dataset_dir='/kaggle/input/strawberry-ripeness-rf-detr-dataset', 
    epochs=3,              # 50 is the "merit" standard for DETR
    batch_size=4,          # Adjust based on Kaggle P100/T4 memory
    grad_accum_steps=4,    # Simulates a batch size of 16
    lr=1e-4,               # Standard fine-tuning learning rat is 1e-4, paired with AdamW Optimizer - but for 1 epoch 2e-4 fits better
    wandb=False            # MUST USE: Shows you're an MLOps pro in your essay
)
```
